```julia
tag = Vector{String}()
NL = Vector{String}()
ind = [2, 1, 3, 4]
for l in L
    if l  [3, 5, 7, 9, 20, 26, 30, 34, 39, 41, 52, 54]
        push!(tag, "Core position")
        push!(NL, "*" * string(l))
    else
        push!(tag, "Remains")
        push!(NL, string(l))
    end
end
data = Pandas.DataFrame(Dict("Feature Importance" => F[ind], "AA type" => tag[ind]));
```

```julia
view_importance(R, NL, F, show_number=30)
data["all"] = ""
set_theme(style="whitegrid", palette="tab10")
swarmplot(x="all", y="Feature Importance", hue="AA type", data=data)
xlabel("")
display(gcf())
close("all")
matplotlib.rc_file_defaults()
```
